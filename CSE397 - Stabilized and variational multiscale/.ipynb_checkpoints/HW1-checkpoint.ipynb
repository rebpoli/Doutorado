```python
import numpy as np
from numpy import tanh
import sys

import matplotlib.pyplot as plt

plt.style.use('paper.mplstyle')
np.set_printoptions(threshold=200, linewidth=200)
plt.figure(figsize=(12,6))

def XI_KB( alpha_h ) :
    global METHOD
    if METHOD == "GALERKIN" :
        return 0
    elif METHOD == "SUPG" :
        return 1/tanh(alpha_h) - 1/alpha_h
    elif METHOD == "UPWIND" :
        return 1
    else :
        fail(f"XI(): Unknown method {METHOD}")

def XI_F( alpha_h ) :
    global METHOD
    if METHOD == "GALERKIN" :
        return 0
    elif METHOD == "SUPG" :
        return 1/tanh(alpha_h) - 1/alpha_h
    elif METHOD == "UPWIND" :
        return 0
    else :
        fail(f"XI(): Unknown method {METHOD}")


def Usolve() :
    global KAPPA, H, A, N, F, G0, G1
    alpha_h = A * H / 2 / KAPPA

    Sdif = np.array( [ -1, 2, -1 ] )
    Sadv = np.array( [ -1, 0, 1] )
    
    K  = np.zeros( [ N-1, N-1 ] )
    B0 = np.zeros( N-1 )
    BN = np.zeros( N-1 )
    FF = np.zeros( N-1 )

    xi = XI_F(alpha_h)
    FF[0]   = F*H/2 * ( 1 - xi )
    FF[N-2] = F*H/2 * ( 1 - xi )    
   
    xi = XI_KB(alpha_h)
    B0[0]   = - ( KAPPA + A*H/2*xi )/H - A/2
    BN[N-2] = - ( KAPPA + A*H/2*xi )/H + A/2

    K[0,0]     = 2*( KAPPA + A*H/2*xi )/H + A
    K[0,1]     = - ( KAPPA + A*H/2*xi )/H + A/2
    K[N-2,N-3] = - ( KAPPA + A*H/2*xi )/H - A/2
    K[N-2,N-2] = 2*( KAPPA + A*H/2*xi )/H - A

    for i in range( 1, N-2 ) :
        [ K[i,i-1],K[i,i],K[i,i+1] ] = ( KAPPA/H + A/2*xi ) * Sdif + A/2 * Sadv

    U = np.linalg.solve( K, FF - B0 * G0 - BN * G1 )
    return U
    
# GALERKIN or SUPG
KAPPA = 1
N = 10
H = 1/N
A = 50
F = 0
G0 = 0
G1 = 1

```

```python

fig, [ax1, ax2, ax3] = plt.subplots( 3,1, figsize=(10,10))

X_EXACT = np.linspace( H, 1, 500 )
U_EXACT = ( np.exp( A*X_EXACT) - 1 ) / ( np.exp(A) - 1)

# NUMERICAL SOLUTIONS
X = np.linspace( H, 1, N-1 )
evr = 1 #int(N/20)

ax=ax1
A=100
U_EXACT = ( np.exp( A*X_EXACT) - 1 ) / ( np.exp(A) - 1)
ax.plot( X_EXACT, U_EXACT, c='k', lw=5, alpha=.3, label='Exact' )
METHOD = "GALERKIN" 
U = Usolve()
ax.plot( X[::evr], U[::evr], marker='x',markeredgewidth=2, ls='--', ms=7, lw=1, label=METHOD, c='blue' )
METHOD = "SUPG" 
U = Usolve()
ax.plot( X[::evr], U[::evr],marker='o', ms=7, ls='-', lw=1, label=METHOD, c='green' )
ax.set_title(f"a={A}")
ax.legend()

ax=ax2
A=10
U_EXACT = ( np.exp( A*X_EXACT) - 1 ) / ( np.exp(A) - 1)
ax.plot( X_EXACT, U_EXACT, c='k', lw=5, alpha=.3, label='Exact' )
METHOD = "GALERKIN" 
U = Usolve()
ax.plot( X[::evr], U[::evr], marker='x',markeredgewidth=2, ls='--', ms=7, lw=1, label=METHOD, c='blue' )
METHOD = "SUPG" 
U = Usolve()
ax.plot( X[::evr], U[::evr],marker='o', ms=7, ls='-', lw=1, label=METHOD, c='green' )
ax.set_title(f"a={A}")
ax.legend()

ax=ax3
A=1
U_EXACT = ( np.exp( A*X_EXACT) - 1 ) / ( np.exp(A) - 1)
ax.plot( X_EXACT, U_EXACT, c='k', lw=5, alpha=.3, label='Exact' )
METHOD = "GALERKIN" 
U = Usolve()
ax.plot( X[::evr], U[::evr], marker='x',markeredgewidth=2, ls='--', ms=7, lw=1, label=METHOD, c='blue' )
METHOD = "SUPG" 
U = Usolve()
ax.plot( X[::evr], U[::evr],marker='o', ms=7, ls='-', lw=1, label=METHOD, c='green' )
ax.set_title(f"a={A}")
ax.legend()

fig.tight_layout()
```
